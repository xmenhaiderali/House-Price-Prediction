```python
#splitting data into train, test
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.05,random_state=0)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
#Standardization of values
scale=StandardScaler().fit(X_train)
X_train_stan=scale.transform(X_train)
X_test_stan=scale.transform(X_test)
```

```python
print(X_train_stan)
```

```python
#Visualization using box plot to determine the effect of StandardScaling on unscaled data
plt.figure(figsize=(10,7))
plt.boxplot(X_train.to_numpy())
plt.show()
```

```python
#Visualization using box plot to determine the effect of StandardScaling on scaled data
plt.figure(figsize=(10,7))
plt.boxplot(X_train.to_numpy())
plt.show()
```

```python
z = np.abs(stats.zscore(X_train))
print(z)
```

```python
threshold = 3
print(np.where(z > 3))
```

```python
rows,columns=np.where(z>3)
rows=list(set(list(rows)))
```

```python
X_train = X_train[(z < 3).all(axis=1)]
```

```python
y_train.drop(y_train.index[rows], inplace=True)
```

```python
X_train.shape,y_train.shape
```

# Linear Regression Model

```python
model_lin_regressor=LinearRegression()
model_lin_regressor.fit(X_train_stan,y_train)
```

```python
y_predict=model_lin_regressor.predict(X_test_stan)
```

```python
print('MSE:', metrics.mean_squared_error(y_test,y_predict))
print('MAE:', metrics.mean_absolute_error(y_test,y_predict))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```

```python
plt.scatter(y_test, y_predict, color='cornflowerblue')
plt.plot([y_test.min(), y_predict.max()], [y_test.min(), y_predict.max()], 'k--', lw=4)
plt.xlabel('Real Price')
plt.ylabel('Predicted')
plt.title("Real vs predicted price")
```

```python
sns.distplot(y_test-y_predict)
```

```python
regressor=RandomForestRegressor()
```

```python
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
print(n_estimators)
```

```python
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 50, num = 10)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10,15,20]
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train_stan,y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
random_forest=RandomForestRegressor(n_estimators=1200,min_samples_split=15, min_samples_leaf=20,max_features='sqrt',max_depth=27)
```

```python
random_forest.fit(X_train,y_train)
y_predict=random_forest.predict(X_test)
```

```python
plt.scatter(y_test, y_predict, color='cornflowerblue')
plt.plot([y_test.min(), y_predict.max()], [y_test.min(), y_predict.max()], 'k--', lw=4)
plt.xlabel('Real Price')
plt.ylabel('Predicted')
plt.title("Real vs predicted price")
```
